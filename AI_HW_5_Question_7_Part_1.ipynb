## <font color='#FF1493'>1. Iris Data Set with Gini
<br>


The code starts by importing the Iris dataset. Then, it assigns an array of shape (150,4) to be array X. This array represents the 4 features (petal length, petal width, sepal length, and sepal width) for all 150 observations. An array of shape (150,) is assigned to be array Y and it represents the labels (Iris setosa, Iris virginica, and Iris versicolor) for each observation. Once arrays X and Y have been assigned, we then instantiate a new decision tree classifier and fit the labels to the features data. By default, the gini coefficient is used as the criterion of the tree. Since the image generated by the plot_tree function was too small, we used the graphviz code to generate a pdf visualization of the decision tree.

```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf)
```

```python
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("iris") 
```

## <font color='#FF1493'>2. Iris Data Set with Entropy
<br>

By default, the gini coefficient is used as the criterion of the tree. To obtain the entropy version of the tree, we manually set entropy as the criterion when the tree was instantiated.

```python
clf_iris2 = tree.DecisionTreeClassifier(criterion='entropy')
clf_iris2 = clf_iris2.fit(X, y)
```

```python
dot_data_iris2 = tree.export_graphviz(clf_iris2, out_file=None) 
graph_iris2 = graphviz.Source(dot_data_iris2) 
graph_iris2.render("iris_entropy") 
```

## <font color='#FF1493'>3. Breast Cancer Data Set with Gini
<br>

The code starts by importing the breast cancer dataset. Then, it assigns an array of shape (569,30) to be array X. This array represents the 30 features (ten features for each cell nucleus: radius, texture, perimeter, area, smoothness, compactness, concavity, concave points, symmetry, and fractal dimension) for all 569 observations. An array of shape (569,) is assigned to be array Y and it represents the labels (M = malignant, B = benign) for each observation. Once arrays X and Y have been assigned, we then instantiate a new decision tree classifier and fit the labels to the features data. By default, the gini coefficient is used as the criterion of the tree. As before, the image generated by the plot_tree function was too small. So, we used the graphviz code to generate a pdf visualization of the decision tree.

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X, y = cancer.data, cancer.target
clf_cancer = tree.DecisionTreeClassifier()
clf_cancer = clf_cancer.fit(X, y)
```

```python
tree.plot_tree(clf_cancer)
```

```python
dot_data_cancer = tree.export_graphviz(clf_cancer, out_file=None) 
graph_cancer = graphviz.Source(dot_data_cancer) 
graph_cancer.render("cancer")
```
